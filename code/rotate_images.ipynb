```python
# Import for rotating
import PIL
from PIL import Image
import os
```

```python
# Data Paths
original_path = "../original_images"
rotate_path = "../rotated_images"
```

```python
original_pictures = os.listdir(original_path)
```

```python
original_pictures
```

```python
j=1
for (i,picture) in enumerate(original_pictures):
    if picture[-3:len(picture)] in ["jpg", "peg"]:
        image = Image.open(os.path.join(original_path, picture))
        if image.size in [(4032,3024), (3024, 4032)]:
            if image.size == (4032,3024):
                image = image.rotate(270, expand = 1)
                print(image.size)
            image.save(os.path.join(rotate_path, "r_image_{iter}.jpg".format(iter=j)))
            j+=1
```
